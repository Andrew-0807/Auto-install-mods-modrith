```python
# Strip mod names
import os
dir = "F:/Stuff/PolyMC-Windows-Portable-5.1/instances/1.20.1/.minecraft/mods"
contents = os.listdir(dir)
print (contents)
```

```python
with open("a.txt", "w") as text:
    for item in contents:
        item = item.replace("-"," ")
        item = item.replace("+"," ")
        item = item.replace(".jar", " ")
        item = item.replace(".", "")
        item = item.replace("mc", "")

        item = ''.join(con for con in item if con.isdigit() != True)

        print(item, "\n")

        text.writelines(item+ "\n")
```
